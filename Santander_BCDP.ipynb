# Extração de Dados - CSV

```python
import pandas as pd

#Seleção do arquivo CSV
df = pd.read_csv("Santander_BCDP_Customers_Information.csv")

#Visualização das 5 primeiras linhas do DataFrame
print(df.head())
```

# Transformação de Dados - OpenAI API

```python
from openai import OpenAI

client = OpenAI()

def gerar_mensagem(row):
    prompt = (
        f"Escreva uma mensagem em PT-BR com no máximo 2 frases, direta e amigável.\n"
        "Não use 'Atenciosamente', não assine com nome, não use palavras em inglês e não use placeholders.\n"
        "Use sempre 'Olá, {nome}!'.\n"
        "Comente saldo e limite de forma útil e educada: se forem baixos, sugira organização/planejamento; se forem altos, sugira aproveitar benefícios com controle.\n"
        "Finalize com uma pergunta curta (sim/não) oferecendo ajuda.\n\n"
        f"Nome: {row['Nome']}\n"
        f"Cartão: {row['Cartao']}\n"
        f"Saldo: R${row['Saldo']}\n"
        f"Limite: R${row['Limite']}\n"
    )

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "Você é um atendente bancário."},
            {"role": "user", "content": prompt}
        ]
    )

    return response.choices[0].message.content.strip()

df["Mensagem"] = df.apply(gerar_mensagem, axis=1)
```

# Carregamento de Dados - CSV

```python
df.to_csv("clientes_com_mensagens.csv", index=False, encoding="utf-8")
```
